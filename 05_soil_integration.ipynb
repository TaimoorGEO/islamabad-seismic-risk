```python
import ee
import urllib.request

# Connect to Earth Engine using your specific project ID
ee.Initialize(project='shm-islamabad-project')

# Define a bounding box roughly covering Islamabad
islamabad_region = ee.Geometry.Rectangle([72.7, 33.5, 73.4, 33.9])

# Grab the updated OpenLandMap USDA soil texture dataset (using the correct TT_M ID)
soil_image = ee.Image("OpenLandMap/SOL/SOL_TEXTURE-CLASS_USDA-TT_M/v02").select('b0')

# Ask Google to generate a direct download link for our specific area and projection
download_url = soil_image.getDownloadURL({
    'scale': 250,          # 250-meter resolution fits perfectly inside our 500m grid
    'crs': 'EPSG:32643',   # Force it into our project's metric coordinate system
    'region': islamabad_region,
    'format': 'GEO_TIFF'
})

print("Downloading soil data directly from Google Earth Engine...")
urllib.request.urlretrieve(download_url, "data/soil_map.tif")
print("Success! GEE soil map saved to data/soil_map.tif.")
```
